# Ejercicio 2: Funciones trigonométricas y exponenciales

```python
import matplotlib.pyplot as plt
```

```python
# Crea un array de valores entre 0 y 2*pi con paso de 0.1

angulos = np.arange(0, 2*np.pi, 0.1)
```

```python
# Calcula los valores de seno, coseno y exponencial para cada ángulo

seno = np.sin(angulos)
coseno = np.cos(angulos)
exponencial = np.exp(angulos)
```

```python
# Grafica las funciones
plt.plot(angulos, seno, label='Seno')
plt.plot(angulos, coseno, label='Coseno')
plt.plot(angulos, exponencial, label='Exponencial')
plt.legend()
plt.title('Funciones Trigonométricas y Exponencial')
plt.xlabel('Ángulo (radianes)')
plt.ylabel('Valor')
plt.show()
```

# Ejercicio 3: Álgebra lineal

```python
# Crea una matriz 3x3 con valores aleatorios entre 1 y 10

matriz = np.random.randint(1,11,size = (3,3))
```

```python
# Calcula la traza de la matriz
traza_matriz = np.trace(matriz)
```

```python
# Calcula la determinante de la matriz
determinante_matriz = np.linalg.det(matriz)
```

```python
# Calcula la matriz inversa (asegúrate de que la matriz sea invertible)
if determinante_matriz !=0:
    matriz_inversa = np.linalg.inv(matriz)

else:
    matriz_inversa = "La Matriz es no Invertible"
```

```python
print(matriz)
```

```python
print(traza_matriz)
```

```python
print(determinante_matriz)
```

```python
print(matriz_inversa)
```
